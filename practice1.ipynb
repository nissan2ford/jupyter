```
# plot data
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('/content/drive/My Drive/Colab Notebooks/8411_2019_mizuho2.csv',encoding='shift-jis', header=0)
date = data["日付"]
endprice = data["終値"]
startprice = data["始値"]

plt.figure(figsize=(20,6))

plt.title("Mizuho Financial Group")
plt.xlabel("Date")
plt.ylabel("End Price[Yen]")

plt.xticks(rotation=90)

plt.plot(date,endprice,marker="o")
plt.plot(date,startprice,marker="x")
plt.show()
```
